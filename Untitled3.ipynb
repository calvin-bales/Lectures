```python
#%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
from numpy import linalg as LA

plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['font.size'] = 16
```

```python
L = 100
W = 500
t = 2

H = np.zeros((L,L))
for i in range(L):
    H[i,i] = (rnd.randint(-W,W))
    if i+1<L:
        H[i+1,i] = t
        H[i,i+1] = t
print(H)
```

```python
v,w = LA.eig(H)
print(np.sort(v))
```

```python
plt.plot(range(L),np.sort(v))
```
