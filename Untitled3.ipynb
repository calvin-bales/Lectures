```python
#%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
from numpy import linalg as LA

plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['font.size'] = 16
```

```python
L = 3
W = 200
t = 2

H = np.zeros((L,L))
for i in range(L):
    H[i,i] = (rnd.randint(-W,W))
    if i+1<L:
        H[i+1,i] = t
        H[i,i+1] = t
print(H)
```

```python
v,w = LA.eig(H)
#print(np.sort(v))
print(w)
```

```python
plt.plot(range(L),np.sort(v), "*", linestyle="-")
plt.grid()
plt.xlabel("n")
plt.ylabel("$E_n$")
```

```python
#many times of the same thing

L = 10
W = 100
t = 2
H = np.zeros((L,L))
N = 100

for j in range(N):
    for i in range(L):
        H[i,i] = (rnd.randint(-W,W))
        if i+1<L:
            H[i+1,i] = t
            H[i,i+1] = t
    v,w = LA.eig(H)
    v+= v/(j+1)

print(np.sort(v))
```
